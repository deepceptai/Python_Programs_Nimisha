Write a Python code snippet to plot a stacked bar chart.

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']

men = [20, 35, 30, 35]

women = [25, 32, 34, 20]

plt.bar(labels, men, color='b', label='Men') 

plt.bar(labels, women, color='r', bottom=men, label='Women')

plt.legend()

plt.show()
```
