Difference between a line plot and a step plot in Matplotlib

A line plot connects data points with a line, while a step plot connects data points with horizontal and vertical lines.

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([1, 3, 4, 5, 7]) 
y = np.array([1, 9, 16, 25, 49]) 

plt.step(x, y, 'g^', where='pre') 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([1, 3, 4, 5, 7]) 
y = np.array([1, 9, 16, 25, 49]) 

plt.step(x, y, 'r*', where='post') 
plt.show() 
```
