The plt.grid() function is used to add a grid to a plot

```python
# Implementation of matplotlib function 
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Plot without grid
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Plot without grid')

# plot with grid
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("Plot with grid")
# draw gridlines
plt.grid(True)

plt.show()
```
