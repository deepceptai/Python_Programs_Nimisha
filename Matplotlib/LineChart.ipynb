Write a Python code snippet to plot a line chart with multiple lines.

```python
import matplotlib.pyplot as plt

import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

y2 = np.cos(x)

plt.plot(x, y1, '-b', label='sin(x)')

plt.plot(x, y2, '--r', label='cos(x)')

plt.legend(loc='upper right')

plt.xlabel('x')

plt.ylabel('y')

plt.show()
```
