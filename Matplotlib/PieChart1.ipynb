Write a Python code snippet to plot a pie chart with different colors for different slices.

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']

sizes = [15, 30, 45, 10]

colors = ['red', 'green', 'blue', 'orange']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')

plt.axis('equal')

plt.show()
```
